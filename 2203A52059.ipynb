```python
import pandas as pd
q1= pd.read_csv("/content/Salary_Data.csv")
print(q1)
```

```python
features = q1[['YearsExperience']]
print(features)

target = q1['Salary']
print(target)
```

```python
import matplotlib.pyplot as plt
plt.scatter(features, target, color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Years of Experience vs Salary')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(features, target)
slope = model.coef_[0]
intercept = model.intercept_
print("Slope:", slope)
print("Intercept:", intercept)
```

```python
plt.scatter(features, target, color='blue')
plt.plot(features, model.predict(features), color='red', linewidth=2)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Years of Experience vs Salary with Best Fit Line')
plt.show()
```
